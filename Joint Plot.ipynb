```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head(4)
```

<br/>
<br/>

## Scatter plot with Reg plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
plt.show()
```

<br/>
<br/>

## Scatter plot in Hexa
##### Hexa is used for observe the density of the data

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hexa' )
plt.show()
```

<br/>
<br/>

## Kde plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde' )
plt.show()
```

<br/>
<br/>

## Scatter plot with Kde plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df).plot_joint(sns.kdeplot, n_levels = 6)
plt.show()
```

<br/>
<br/>

## Added Rug plot also

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df).plot_joint(sns.kdeplot, n_levels = 6).plot_marginals(sns.rugplot)
plt.show()
```

<br/>
<br/>

## Stat Function

```python
from scipy.stats import spearmanr
sns.jointplot(x = 'total_bill', y = 'tip', data = df, stat_func = spearmanr)
plt.show()
```
