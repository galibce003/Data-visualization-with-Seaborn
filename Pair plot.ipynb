```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df.head()
```

<br/>
<br/>

## A primary pair plot

```python
sns.pairplot(df)
plt.show()
```

<br/>
<br/>

## Pair plot with kde plot in diagonal

```python
sns.pairplot(df, diag_kind= 'kde')
plt.show()
```

<br/>
<br/>

## Pair plot with hue

```python
sns.pairplot(df, hue = 'species')
plt.show()
```

<br/>
<br/>

## Pairplot with marker

```python
sns.pairplot(df, hue = 'species', markers = ['o', 's', 'D'])
plt.show()
```

<br/>
<br/>

## Pair plot with regression line

```python
sns.pairplot(df, kind= 'reg')
plt.show()
```

<br/>
<br/>

## Bring all together

```python
sns.pairplot(df, kind = 'reg', diag_kind= 'kde', hue = 'species')
plt.show()
```

<br/>
<br/>

## Pair plot of two selected variables

```python
sns.pairplot(df, hue = 'species', vars = ['sepal_width', 'sepal_length'])
plt.show()
```

<br/>
<br/>

## Also select the axis

```python
sns.pairplot(df, hue = 'species',
            x_vars = ['sepal_width', 'sepal_length'],
            y_vars = ['petal_width', 'petal_length'])
plt.show()
```
