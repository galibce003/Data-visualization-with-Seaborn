```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head(4)
```

<br/>
<br/>

## Primary Regression plot

```python
sns.regplot(x='total_bill', y='tip', data = df)
plt.show()
```

<br/>
<br/>

## Regression plot with customized data point and trend line

```python
sns.regplot(x='total_bill', y='tip', data = df,
            scatter_kws = {'marker' : 'D', 'color' : 'red', 'edgecolor' : 'black', 'alpha' : 0.5},
            line_kws = {'color' : 'purple', 'linewidth' :4.5})
plt.show()
```

<br/>
<br/>

## Regression plot with confidence interval

```python
sns.regplot(x='total_bill', y='tip', data = df, ci = 95)
plt.show()
```

<br/>
<br/>

## Regression plot with x_estimator

```python
sns.regplot(x='size', y='tip', data = df, x_estimator = np.mean)
plt.show()
```
