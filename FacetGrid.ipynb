```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = sns.load_dataset('tips')
df.head(4)
```

<br/>
<br/>

## Primary FacetGrid

```python
fg = sns.FacetGrid(df, col = 'sex', row = 'smoker')
fg = fg.map(sns.distplot, 'total_bill')
```

<br/>
<br/>

## FacetGrid with ScatterPlot

```python
fg = sns.FacetGrid(df, col = 'time', row = 'smoker')
fg = fg.map_dataframe(sns.scatterplot, 'total_bill', 'tip', size = 'size', color = 'red', alpha = 0.5)
```

<br/>
<br/>

## FacetGrid with customized order

```python
fg = sns.FacetGrid(df, col = 'smoker', col_order = ['No', 'Yes'])
fg = fg.map(sns.distplot, 'total_bill')
```

<br/>
<br/>

## Customize the axis label

```python
fg = sns.FacetGrid(df, col = 'smoker', row = 'sex')
fg = (fg.map(sns.scatterplot, 'total_bill', 'tip').set_axis_labels('Total Bill (US)', 'Tips'))
```

<br/>
<br/>

## Wrap the column

```python
fg = sns.FacetGrid(df, col = 'size', col_wrap = 3)
fg = (fg.map(sns.distplot, 'tip', kde = False).set_titles('{col_name} dinners'))
```

<br/>
<br/>

## Without wraping the columns

```python
fg = sns.FacetGrid(df, col = 'size')
fg = (fg.map(sns.distplot, 'tip', kde = False).set_titles('{col_name} dinners'))
```
