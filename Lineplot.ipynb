## Lineplot is used for observing the relations between two numerical variables
<br/>
<br/>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head(4)
```

<br/>
<br/>

## Primary lineplot

```python
sns.lineplot(x='total_bill', y='tip', data = df)
plt.show()
```

<br/>
<br/>

## Lineplot grouped by one variable and separated by color

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='total_bill', y='tip', hue= 'sex', data = df)
plt.show()
```

<br/>
<br/>

## Lineplot grouped by two variables and separated by color and style

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='total_bill', y='tip',hue= 'sex', style = 'smoker', data = df)
plt.show()
```

<br/>
<br/>

## Lineplot grouped by two variables and separated by color and linewidth

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='total_bill', y='tip',hue= 'sex', size = 'smoker', data = df)
plt.show()
```
