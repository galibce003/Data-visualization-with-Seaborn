## Scatterplot is used for observing the relations between two numerical variables
<br/>
<br/>

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head(4)
```

<br/>
<br/>

## Primary Scatterplot

```python
sns.scatterplot(x='total_bill', y='tip', data = df)
plt.show()
```

<br/>
<br/>

## Scatterplot grouped by one variable and separated by color

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='total_bill', y='tip', hue= 'sex', data = df)
plt.show()
```

<br/>
<br/>

## Scatterplot grouped by two variables and separated by color and style

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='total_bill', y='tip', hue= 'sex', style = 'smoker', data = df, s = 100)
plt.show()
```

<br/>
<br/>

## Scatterplot grouped by two variables and separated by color and size

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='total_bill', y='tip', hue= 'sex', size = 'smoker', data = df, s = 100)
plt.show()
```
